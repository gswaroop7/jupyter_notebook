```python
from pyspark import SparkContext
```

```python
sc=SparkContext("local","wordCount")
```

```python
rdd=sc.textFile("log.txt")
```

```python
rdd1=rdd.flatMap(lambda x: x.split(" "))  #or check len(x) > 0
```

```python
rdd2=rdd1.filter(lambda x: x != '')
```

```python
rdd2.map(lambda x: (x,1)).reduceByKey(lambda x,y: x+y).sortBy(lambda x:x[1],False).collect()
```
